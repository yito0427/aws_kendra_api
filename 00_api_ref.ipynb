## query
https://boto3.amazonaws.com/v1/documentation/api/latest/reference/services/kendra/client/query.html

```python
client.query(
    IndexId='0890c7c0-6918-46bf-ad70-b0206e00ad47',
    QueryText='s3',
    AttributeFilter={
        'EqualsTo': {
            'Key': '_language_code',
            'Value': {
                'StringValue': 'ja',
            }
        },
    }
)
```
